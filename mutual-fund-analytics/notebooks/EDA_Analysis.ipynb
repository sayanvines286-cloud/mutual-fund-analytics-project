```python
import sqlite3
import pandas as pd

conn = sqlite3.connect("../bluestock_mf.db")

nav = pd.read_sql("SELECT * FROM nav_history", conn)
transactions = pd.read_sql("SELECT * FROM transactions", conn)
performance = pd.read_sql("SELECT * FROM performance", conn)

print(nav.head())
print(transactions.head())
print(performance.head())
```

```python
import pandas as pd
import plotly.express as px

# Convert date column
nav["date"] = pd.to_datetime(nav["date"])

# Average NAV per day
daily_nav = nav.groupby("date")["nav"].mean().reset_index()

fig = px.line(
    daily_nav,
    x="date",
    y="nav",
    title="Average Daily NAV (2022–2026)"
)

fig.show()
```
